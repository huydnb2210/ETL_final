## Save as DataFrame

```python
#Convert data as dictionary using request.json()
data = request.json()
#Turn the data into a dataframe
dataframe = pd.DataFrame.from_dict(data)
```

```python
dataframe
```

```python
#Convert timestamp as real time
from datetime import datetime
dataframe['timestamp'] = dataframe['timestamp'].apply(lambda d: datetime.fromtimestamp(d/1000.0))
dataframe
```

```python
#Only show currency and rates
dataframe = dataframe.drop(['success','timestamp' , 'base', 'date'], axis = 1)
dataframe
```

## Load the data
